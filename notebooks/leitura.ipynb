```python
import pandas as pd
import glob #procurar arquivos
```

```python
path = "/home/holyvieri/think-learn-ai/data/raw"
path_pattern = path + "/*.parquet.gzip"

lista_paths = glob.glob(path_pattern)
lista_paths

lista_dfs = []
for file in lista_paths:
    df = pd.read_parquet(file)
    lista_dfs.append(df)
```

```python
df_appended = pd.concat(lista_dfs, ignore_index=True)
```

```python
path_to_save = "/home/holyvieri/think-learn-ai/data/processed/"
df_appended.to_parquet(path_to_save+"df_append.parquet.gzip", compression='gzip')
```
